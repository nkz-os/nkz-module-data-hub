## Consultar series temporales

La función `get_dataframe` descarga datos alineados en formato Apache Arrow y los convierte en un `pandas.DataFrame` de forma eficiente.

```python
# Ejemplo: consultar humedad del suelo y temperatura de un sensor IoT
# Sustituye device_id por un ID real de tu tenant

df = await nkz.timeseries.get_dataframe(
    device_id="120786a0cf364796",
    attributes=["soilMoisture", "airTemperature"],
    start_date="2026-03-01",
    end_date="2026-03-28",
    resolution=500,
)

df.head()
```
